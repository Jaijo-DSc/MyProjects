```python
import requests
import json
import pandas as pd

def clean_data(data):
    cleaned_data = {}
    for key, value in data.items():
        if value is not None:
            cleaned_data[key] = value
        else:
            print ("Null values removing from dict")
    return cleaned_data

def main(celebrities):
    url = "https://api.fullcontact.com/v3/person.enrich"
    api_key = "enter api"
    processed_data = []
    for celebritie in celebrities:
        payload = {
            "twitter": celebritie
        }
        headers = {
            "Authorization": "Bearer " + api_key,
            "Content-Type": "application/json"
        }
        response = requests.request("POST", url, data=json.dumps(payload), headers=headers)
        informations = response.json()
        relevent_information = {
            "full_name": informations["fullName"],
            "gender": informations["gender"],
            "location": informations["location"],
            "title": informations["title"],
            "twitter_link": informations["twitter"],
            "facebook": informations["facebook"]
        }
        cleaned_data = clean_data(relevent_information)
        processed_data.append(cleaned_data)

    # print (processed_data)
    df = pd.DataFrame(processed_data)
    print (df)

if __name__== "__main__":
  celebrities = ["realDonaldTrump", "narendramodi", "SrBachchan", "sachin_rt","shahrukhkhan","JohnnyDepp"," jimmyfallon","BillGates","SriDevi","GerardButler","willsmith","JuhiChawla","AkshayKumar","ViratKohli","SachinTendulkar","EmmaWatson","Kaka","HrithikRoshan","kanganaRanaut","LionelMessi","Priyanka","RickyMartin","AngelinaJolie","Beyonce","AudreyHepburn"] # you can add more celebrities here.
  print ("Please wait... getting the data")
  main(celebrities)




```
