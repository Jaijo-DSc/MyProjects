```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.read_excel(" Australia_Filtered.xlsx")
```

# 1. Plot bar chart for top 5 countires having max no of total emigrants

```python
r=df.sort_values(by = "Total").tail()
```

```python
r
```

```python
r.plot(x = "Country", y = "Total",kind="bar")
plt.show()
```

```python
#  matplotlib
r.plot(x = "Country", y = "Total",kind="bar")
plt.show()


print()
# matplotlib.pyplot
plt.bar(r.Country.values, r.Total.values)
plt.xticks(rotation =90)
```

# 2.  Find total Emigrants from each category by area name and plot barchart for each area

```python
s = df.groupby("AreaName")
```

```python
t =s.Total.sum()
```

```python
t
```

```python
#  matplotlib
t.plot(x= "AreaName", y="Total",kind ="bar")
plt.show()

print()
# matplotlib.pyplot
plt.bar(t.index,t.values)
plt.xticks(rotation=90)
plt.show()



```

# 3. Plot total Emigration trend(Emigration of world over the years i.e. year vs Total count for that year) using line plot.

```python
#  matplotlib
Q = df.iloc[:,5:-1].sum()

print(Q)
#print(Q.to_frame())
#print(Q.to_frame().T)# for tabular formate
Q.plot(kind = 'line')
plt.show()

print()
```

```python
# matplotlib.pyplot

plt.plot(Q.index,Q.values)
plt.show()
```

# 4. Compare Emigration Trend for Asia, Africa and Europe AreaName (Use Line plot).

```python
df.head()
```

```python
z =df.groupby("AreaName")
```

```python
Asia = z.get_group("Asia").iloc[:,5:-1].sum()
plt.plot(Asia,label = "Asia")
Africa = z.get_group("Africa").iloc[:,5:-1].sum()
plt.plot(Africa,label = "Africa")
Europe = z.get_group("Europe").iloc[:,5:-1].sum()
plt.plot(Europe,label = "Europe")
plt.legend()
plt.xticks(rotation =90)
plt.show()
```

# 5  Plot Total Emigrants based on DevelopmenRegion


```python
df

```

```python
# Plot Total Immigrants based on DevelopmenRegion 

plt.figure(figsize=(16,5))


years = df.iloc[: ,5:-1]
years =list(years)


res = df.groupby('AreaName')[years].sum() # group by area and take out years data
print(res)

for row in res.index:
    plt.plot(res.loc[row], label = row)
    
plt.xticks(rotation = 90)
plt.legend()
plt.show()


```

# 6 Plot Emigration trend for India using line plot


```python
idx = df.index[df.Country == "India"]
data =df.loc[idx].iloc[:,5:-1].T
data.plot(kind="line")
plt.show()
```

# 7. Plot Emigration Trend(year vs count) for the coutries with top 5 total Emigrants. Use Line plot

```python
# Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot

plt.figure(figsize=(16,5))
data = df.sort_values(by = 'Total', ascending= False).head()
label = data.Country
values = data.iloc[:,5:-1]


df1 = pd.DataFrame(values.T)
df1.columns = label


df1.plot(kind = 'line')
plt.legend(fontsize=5)
plt.show()
```

# 8 For the above use area plot with alpha = 0.5

```python
plt.figure(figsize=(16,5))
data = df.sort_values(by = 'Total', ascending= False).head()
label = data.Country
values = data.iloc[:,5:-1]

# build a dataframe to represent Data
df1 = pd.DataFrame(values.T)
df1.columns = label


df1.plot(kind = 'area', stacked = False, alpha = 0.5)
plt.legend(fontsize=5)
plt.show()
```

# 9 Compare Emigration Trend for top-5 countires using stacked area plot

```python

plt.figure(figsize=(16,5))
data = df.sort_values(by = 'Total', ascending= False).head()
label = data.Country
values = data.iloc[:,5:-1]


df1 = pd.DataFrame(values.T)
df1.columns = label

df1.plot(kind = 'area', stacked = True)
plt.legend(fontsize=5)
plt.show()
```
