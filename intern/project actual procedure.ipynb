```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(x_o)
```

```python
###### splitting the model
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# model training 
dtc = DecisionTreeClassifier(criterion="gini", max_depth=10,min_samples_leaf=5)
dtc.fit(X_train,y_train)
```

```python
y_predict  = dtc.predict(X_test)
```

```python
accuracy_score(y_predict,y_test)
```

```python
# fitting the model
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=6)
model.fit(X_train,y_train)
#predicting the value with X_test
y_pre = model.predict(X_test)
```

```python
# Accuracy
accuracy_score(y_pre,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
```

```python
gnb.score(X_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
# model fiting
LogR =LogisticRegression(solver ="saga", multi_class='ovr')
LogR.fit(X_train,y_train)
```

```python
#predict the new value of X_test

y_predict = LogR.predict(X_test)
```

```python
# Evalution metrics
print(classification_report(y_test,y_predict))
```

```python
from sklearn.svm import SVC
#model training
model = SVC(C=0.3,kernel ="rbf", )
model.fit(X_train,y_train)
#predicting the value
y_pred =model.predict(X_test)
```

```python
# score
accuracy_score(y_pred,y_test)
```
